# Predictive Sequential Evaluation on Covtype using Scikit-Multiflow

```python
%matplotlib notebook
from skmultiflow.data.file_stream import FileStream
from skmultiflow.trees.hoeffding_tree import HoeffdingTree
from skmultiflow.meta.adaptive_random_forests import AdaptiveRandomForest
from skmultiflow.lazy.knn_adwin import KNNAdwin
from skmultiflow.evaluation.evaluate_prequential import EvaluatePrequential
```

## Generate stream

```python
stream = FileStream('~/data/covtype/covtype.csv') # taken from the sk-multiflow repository
stream.prepare_for_use()
stream.n_samples
```

## Hoeffding Tree

```python
model = HoeffdingTree()
evaluator = EvaluatePrequential(show_plot=True, pretrain_size=200, max_samples=20000)
evaluator.evaluate(stream=stream, model=model)
```

## Adaptive Random Forest

```python
model = AdaptiveRandomForest()
evaluator = EvaluatePrequential(show_plot=True, pretrain_size=200, max_samples=20000)
evaluator.evaluate(stream=stream, model=model)
```

## SAM KNN

```python
model = KNNAdwin()
evaluator = EvaluatePrequential(show_plot=True, pretrain_size=200, max_samples=20000)
evaluator.evaluate(stream=stream, model=model)
```
